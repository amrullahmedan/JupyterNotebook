```python
import matplotlib.pyplot as plt 
import pandas as pd
ds = pd.read_csv('covid19.csv')
ds
```

# Kasus covid19 di eropa 

```python
kasusger = ds[ds['Country/Region'] == 'Germany']
kasusger
```

```python
kasusger1 = kasusger[kasusger['Confirmed']>1]
kasusger1
```

```python
kasusger1['Date'] = pd.to_datetime(kasusger1['Date'])
kasusger1
```

```python
kasusger1 ['BaseDate'] =  pd.to_datetime('2020-01-2')
kasusger1
```

```python
kasusger1['Harike'] = kasusger1['Date'] - kasusger1['BaseDate']
kasusger1.head()
```

```python
kasusger1.Harike = kasusger1.Harike/pd.Timedelta(1, unit='d')
kasusger1
```

#  Kasus covid di benua Asia

```python
kasusjpn = ds[ds['Country/Region'] == 'Japan']
kasusjpn
```

```python
kasusjpn1 = kasusjpn[kasusjpn['Confirmed']>0]
kasusjpn1
```

```python
kasusjpn1['Date'] = pd.to_datetime(kasusjpn1['Date'])
kasusjpn1
```

```python
kasusjpn1 ['BaseDate'] =  pd.to_datetime('2020-01-22')
kasusjpn1
```

```python
kasusjpn1['Harike'] = kasusjpn1['Date'] - kasusjpn1['BaseDate']
kasusjpn1.head()
```

```python
kasusjpn1.Harike = kasusjpn1.Harike/pd.Timedelta(1, unit='d')
kasusjpn1.head()
```

```python
plt.plot(kasusger1.Harike, kasusger1.Confirmed)
plt.plot(kasusjpn1.Harike, kasusjpn1.Confirmed)
plt.xlabel('Hari Ke-')
plt.ylabel('Jumlah Kasus')
plt.legend('Germany, Japan')
plt.title('jumlah Perbandingan Kasus Covid19 Ger & Jpn')
```
