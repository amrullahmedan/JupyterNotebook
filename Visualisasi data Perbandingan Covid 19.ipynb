```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('covid19.csv')
df
```

```python
kasus1 =df[df['Country/Region'] == 'Indonesia']
kasus1
```

```python
kasusafg = df[df['Country/Region'] == 'Afghanistan']
kasusafg
```

```python
kasusalb = df[df['Country/Region'] == 'Albania']
kasusalb 
```

```python
kasusafg = kasusafg[kasusafg['Confirmed']>0]
kasusafg
```

```python
kasusalb = kasusalb[kasusalb['Confirmed']>0]
kasusalb
```

```python
kasusind['Date'] = pd.to_datetime(kasusind['Date'])
kasusalb['Date'] = pd.to_datetime(kasusalb['Date'])
kasusafg['Date'] = pd.to_datetime(kasusafg['Date'])
kasusafg.head()
```

```python
kasusind['BaseDate'] = pd.to_datetime('2020-03-02')
kasusalb['BaseDate'] = pd.to_datetime('2020-03-09')
kasusafg['BaseDate'] = pd.to_datetime('2020-02-24')
kasusafg
```

```python
kasusind['Harike'] = kasusind['Date'] - kasusind['BaseDate']
kasusalb['Harike'] = kasusalb['Date'] - kasusalb['BaseDate']
kasusafg['Harike'] = kasusafg['Date'] - kasusafg['BaseDate']
kasusind.head()
```

```python
kasusind.Harike = kasusind.Harike/pd.Timedelta(1, unit='d')
```

```python
kasusind.head()
```

```python
kasusafg.Harike = kasusafg.Harike/pd.Timedelta(1, unit='d')
```

```python
kasusafg
```

```python
kasusalb.Harike = kasusalb.Harike/pd.Timedelta(1, unit='d')
```

```python
kasusalb.head()
```

```python
plt.plot(kasusind.Harike, kasusind.Confirmed)
plt.plot(kasusalb.Harike, kasusalb.Confirmed)

plt.xlabel('Hari Ke-')
plt.ylabel('Jumlah Kasus')
plt.legend(['Indonesia, Albania'])
plt.title('Jumlah Perbandingan Covid 19 ')
```
