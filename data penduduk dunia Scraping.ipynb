```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url ='https://id.wikipedia.org/wiki/Daftar_bank_di_Indonesia'
UserAgent = Request(url,headers={'User-Agent':'Mozilla/5.0'})
html_text = urlopen(UserAgent)
soup = BeautifulSoup(html_text,'html.parser')
```

```python
tabel = soup.findAll('table',{'class':'wikitable'})[1]
baris = tabel.findAll('tr')
```

```python
for row in baris:
    for cell in row.findAll(['td','th']):
        cell.get_text()
```

```python
tampildata = []
for row in baris:
    info =[]
    for cell in row.findAll(['td','th']):
        info.append(cell.get_text().strip())
    tampildata.append(info)
```

```python
df = pd.DataFrame(tampildata)
display(df)
```

```python
df = pd.DataFrame(tampildata)
new_header = df.iloc[0]
df = df[1:]
df.columns = new_header
df.reset_index(drop = True, inplace = True)
display(df)
```

```python
df.to_csv('kodebanknegara.csv')
```
