```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('covid19.csv')
df
```

```python
kasus1 = df[df['Country/Region'] == 'Indonesia']
kasus1
```

```python
kasusmy = df[df['Country/Region'] == 'Malaysia']
kasusmy
```

```python
kasusind = kasus1[kasus1['Confirmed']>0]
kasusind
```

```python
kasusmy = kasusmy[kasusmy['Confirmed']>0]
kasusmy
```

```python
kasusind['Date'] = pd.to_datetime(kasusind['Date'])
kasusmy['Date'] = pd.to_datetime(kasusmy['Date'])
kasusind.head()
```

```python
kasusind['BaseDate'] = pd.to_datetime('2020-03-02')
kasusmy['BaseDate'] = pd.to_datetime('2020-01-25')
kasusmy.head()
```

```python
kasusind['HariKe'] = kasusind['Date'] - kasusind['BaseDate']
kasusmy['HariKe'] = kasusmy['Date'] - kasusmy['BaseDate']
kasusind.head()
```

```python
kasusind.HariKe = kasusind.HariKe/pd.Timedelta(1, unit='d')
```

```python
kasusind.head()
```

```python
kasusmy.HariKe = kasusmy.HariKe/pd.Timedelta(1, unit='d')
```

```python
kasusmy.head()
```

```python
plt.plot(kasusind.HariKe, kasusind.Confirmed)
plt.plot(kasusmy.HariKe, kasusmy.Confirmed)
plt.xlabel('Hari Ke-')
plt.ylabel('jumlah Kasus')
plt.legend(['Indonesia', 'Malaysia'])
plt.title('Jumlah perbandingan kasus')
plt.show()
```
